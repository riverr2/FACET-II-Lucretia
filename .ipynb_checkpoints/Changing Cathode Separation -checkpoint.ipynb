```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('/u/gu/riverr/PaperDoubleFig.mplstyle')
from scipy.io import loadmat
import subprocess as sp
import matlab.engine
```

```python
def run_mat(P1, P2, dt, fac):
    with open('twobunch_scan_fcn_newS20_v3.m', 'r') as f:
        lines = f.readlines()
    newlines = []
    for line in lines:
        if line.startswith('load beam'):
            newlines.append(f'load beam_{dt}_dt_{fac}_fac.mat\n')
        else:
            newlines.append(line)
    
    with open('twobunch_scan_fcn_newS20_v3.m', 'w') as f:
        f.writelines(newlines)
    
    with open('baseline_sim.m', 'r') as f:
        lines = f.readlines()
    
    newlines = []
    for line in lines:
        if line.startswith('P1'):
            newlines.append(f'P1 = {P1};\n')
        elif line.startswith('P2'):
            newlines.append(f'P2 = {P2};\n')
        else:
            newlines.append(line)
    
    with open('new_sim.m', 'w') as f:
        f.writelines(newlines)
        
    eng = matlab.engine.start_matlab()
    eng.new_sim(nargout=0)
    eng.quit()
    

def process_output(bc, nbins):
    dat = loadmat(f'{bc}.mat', simplify_cells=True)[bc]
    wit_idx = (dat['Q']==np.min(np.unique(dat['Q'])))&(dat['stop']==0)
    drive_idx = (dat['Q']==np.max(np.unique(dat['Q'])))&(dat['stop']==0)
    bunch_sep = np.abs(np.mean(dat['x'][4,drive_idx]) - np.mean(dat['x'][4,wit_idx]))
    zs = np.linspace(dat['x'][4,drive_idx].min(), dat['x'][4,drive_idx].max(), nbins)
    stop_idx = dat['stop']==0
    #plt.plot(dat['x'][4,stop_idx], dat['x'][5,stop_idx], '.')
    #plt.show()
    curs = []
    dz = np.mean(np.diff(zs))
    for i, z in enumerate(zs):
        new_idx = (dat['x'][4,drive_idx]>zs.min()+i*dz)&(dat['x'][4,drive_idx]<=zs.min()+(i+1)*dz)
        curs.append(np.sum(dat['Q'][drive_idx][new_idx])/(dz/3e8))
    drive_cur = np.max(curs)
    
    zs = np.linspace(dat['x'][4,wit_idx].min(), dat['x'][4,wit_idx].max(), nbins)
    stop_idx = dat['stop']==0
    #plt.plot(dat['x'][4,stop_idx], dat['x'][5,stop_idx], '.')
    #plt.show()
    curs = []
    dz = np.mean(np.diff(zs))
    for i, z in enumerate(zs):
        new_idx = (dat['x'][4,wit_idx]>zs.min()+i*dz)&(dat['x'][4,wit_idx]<=zs.min()+(i+1)*dz)
        curs.append(np.sum(dat['Q'][wit_idx][new_idx])/(dz/3e8))
    zmean = np.sum(zs*curs)/np.sum(curs)
    zrms = np.sqrt(np.sum((zs-zmean)**2*curs)/np.sum(curs))
    return bunch_sep, drive_cur, zrms
```

```python
np.seterr(all='raise')

def get_slice_dat(dt, fac, nbins):
    dat = loadmat(f'beam_{dt}_dt_{fac}_fac.mat', simplify_cells=True)['end_inj']['Bunch']
    wit_idx = (dat['Q']==np.min(np.unique(dat['Q'])))&(dat['stop']==0)
    drive_idx = (dat['Q']==np.max(np.unique(dat['Q'])))&(dat['stop']==0)
    bunch_sep = np.abs(np.mean(dat['x'][4,drive_idx]) - np.mean(dat['x'][4,wit_idx]))
    stop_idx = dat['stop']==0
    idx = stop_idx
    zs = np.linspace(dat['x'][4,idx].min(), dat['x'][4,idx].max(), nbins)

    #plt.plot(dat['x'][4,stop_idx], dat['x'][5,stop_idx], '.')
    #plt.show()
    curs_drive = []
    curs_wit = []
    emits_drive = []
    emits_wit = []
    dz = np.mean(np.diff(zs))
    for i, z in enumerate(zs):
        idx = drive_idx
        new_idx = (dat['x'][4,idx]>zs.min()+i*dz)&(dat['x'][4,idx]<=zs.min()+(i+1)*dz)
        curs_drive.append(np.sum(dat['Q'][idx][new_idx])/(dz/3e8))
        x = dat['x'][0,idx][new_idx]
        xp = dat['x'][1,idx][new_idx]
        try:
            emit = (np.mean(dat['x'][5,idx][new_idx])/0.511e-3*np.sqrt(x.std()**2*xp.std()**2 - np.mean((x-x.mean())*(xp-xp.mean()))**2))
        except FloatingPointError as e:
            emit = 0
        emits_drive.append(emit)
        idx = wit_idx
        new_idx = (dat['x'][4,idx]>zs.min()+i*dz)&(dat['x'][4,idx]<=zs.min()+(i+1)*dz)
        curs_wit.append(np.sum(dat['Q'][idx][new_idx])/(dz/3e8))
        x = dat['x'][0,idx][new_idx]
        xp = dat['x'][1,idx][new_idx]
        try:
            emit = (np.mean(dat['x'][5,idx][new_idx])/0.511e-3*np.sqrt(x.std()**2*xp.std()**2 - np.mean((x-x.mean())*(xp-xp.mean()))**2))
        except FloatingPointError as e:
            emit = 0
        emits_wit.append(emit)
    curs_drive = np.array(curs_drive)
    curs_wit = np.array(curs_wit)
    emits_drive = np.array(emits_drive)
    emits_wit = np.array(emits_wit)
    
    return zs, curs_drive, curs_wit, emits_drive, emits_wit, np.sum(curs_drive*emits_drive)/np.sum(curs_drive), np.sum(curs_wit*emits_wit)/np.sum(curs_wit)
```

```python
res = get_slice_dat(90, 10, 500)

plt.plot(res[0], res[4])
```

```python
dts = np.array([30, 40, 50, 60, 70, 80, 90])
fac = 25
emits_dr = []
emits_w = []
for dt in dts:
    res = get_slice_dat(dt, fac, 500)
    emits_dr.append(res[-2])
    emits_w.append(res[-1])
emits_dr = np.array(emits_dr)
emits_w = np.array(emits_w)
```

```python
plt.plot(dts*11/100, 1e6*emits_dr, label='Drive')
plt.plot(dts*11/100, 1e6*emits_w, label='Witness')
plt.xlabel('Cathode Separation (ps)')
plt.ylabel(r'Avg. Norm. Slice Emittance ($\mu$m)')
plt.legend()
plt.title(f'{500/fac} pC Witness Charge')
plt.ylim(0, 3)
plt.savefig(f'{500/fac}_pC_witness_emits.png', dpi=500)
```
